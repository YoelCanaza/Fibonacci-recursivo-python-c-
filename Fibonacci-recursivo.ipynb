```python
import time
import matplotlib.pyplot as plt


def fibonacci_recursivo(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)


def medir_tiempo_fibonacci(n):
    start_time = time.time()
    fibonacci_recursivo(n)
    end_time = time.time()
    return end_time - start_time


n_values = list(range(1, 35))  # Analizaremos hasta el término 34 por motivos de tiempo


tiempos_ejecucion = [medir_tiempo_fibonacci(n) for n in n_values]


plt.figure(figsize=(10, 6))
plt.plot(n_values, tiempos_ejecucion, marker='o', linestyle='-', color='b')
plt.title('Tiempo de Ejecución de Fibonacci Recursivo')
plt.xlabel('Término enésimo (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.grid(True)
plt.show()
```
